program 1 : A python program to display distribution plot by taking number if cylinders on x axis.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data Creation for Demonstration
data = {
    'cylinders': [4, 6, 4, 8, 6, 4, 6, 8, 4, 6, 4, 8, 6, 4],
    'values': [30, 45, 35, 60, 50, 40, 48, 70, 38, 52, 33, 65, 47, 36]
}

df = pd.DataFrame(data)

# Plotting the Distribution Plot
sns.histplot(data=df, x='cylinders',bins=4 )

# Adding Title and Labels
plt.title("Distribution Plot by Number of Cylinders", fontsize=16)
plt.xlabel("Number of Cylinders", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Display the Plot
plt.show()
```

Program 2 :  A python program to display distribution plot by taking number of cylinders on x axis and mileage on y axis 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data Creation for Demonstration
data = {
    'cylinders': [4, 6, 4, 8, 6, 4, 6, 8, 4, 6, 4, 8, 6, 4],
    'mileage': [30, 25, 32, 15, 22, 35, 20, 12, 33, 24, 31, 14, 23, 34]
}

df = pd.DataFrame(data)

# Plotting the Distribution Plot
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='cylinders', y='mileage', cmap="viridis")

# Adding Title and Labels
plt.title("Distribution Plot: Cylinders vs Mileage", fontsize=16)
plt.xlabel("Number of Cylinders", fontsize=12)
plt.ylabel("Mileage (MPG)", fontsize=12)

# Display the Plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data Creation for Demonstration
data = {
    'cylinders': [4, 6, 4, 8, 6, 4, 6, 8, 4, 6, 4, 8, 6, 4],
    'mileage': [30, 25, 32, 15, 22, 35, 20, 12, 33, 24, 31, 14, 23, 34]
}

df = pd.DataFrame(data)

# Plotting the KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='cylinders', y='mileage', cmap="viridis", fill=True)

# Adding Title and Labels
plt.title("Distribution Plot: Cylinders vs Mileage", fontsize=16)
plt.xlabel("Number of Cylinders", fontsize=12)
plt.ylabel("Mileage (MPG)", fontsize=12)

# Display the Plot
plt.show()
```
