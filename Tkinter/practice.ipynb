```python
import tkinter as tk

# Create the main window
window = tk.Tk()
window.title("Row and Column Example")
window.geometry("400x400")  # Set the size of the window

# Create and place labels in a grid
for row in range(4):  # Loop through rows
    for col in range(3):  # Loop through columns
        label = tk.Label(window, text=f"Row {row}, Col {col}", font=("Arial", 12), bg="lightblue")
        label.grid(row=row, column=col, padx=10, pady=10, sticky="nsew")  # Add spacing and alignment

# Configure row and column weights (optional for resizing behavior)
for i in range(5):  # Number of rows
    window.grid_rowconfigure(i, weight=1)
for j in range(3):  # Number of columns
    window.grid_columnconfigure(j, weight=1)

# Run the application
window.mainloop()
```

```python
import tkinter as tk

# Create the main window
window = tk.Tk()
window.title("Menu Example")
window.geometry("400x300")

# Create a menu bar
menu = tk.Menu(window)

# Create a "File" drop-down menu
file = tk.Menu(menu, tearoff=0)  # `tearoff=0` makes the menu fixed
file.add_command(label="New", command=lambda: print("New File"))
file.add_command(label="New Window", command=lambda: print("New Window"))
file.add_command(label="Open", command=lambda: print("Open File"))
file.add_command(label="Save", command=lambda: print("Save File"))
file.add_command(label="Save As", command=lambda: print("Save As File"))
file.add_separator()  # Add a separator line
file.add_command(label="Exit", command=window.quit)

# Add the "File" menu to the menu bar
menu.add_cascade(label="File", menu=file)

# Add more menus (optional)
edit = tk.Menu(menu, tearoff=0)
edit.add_command(label="Undo", command=lambda: print("Undo"))
edit.add_command(label="Redo", command=lambda: print("Redo"))
menu.add_cascade(label="Edit", menu=edit)

# Configure the menu bar
window.config(menu=menu)

# Run the application
window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

# Create the main window
window = tk.Tk()
window.title("Messagebox Example")
window.geometry("300x200")

# Function to show an info message
def show_info():
    messagebox.showinfo("Information", "This is an informational message.")

# Function to show a warning message
def show_warning():
    messagebox.showwarning("Warning", "This is a warning message.")

# Function to show an error message
def show_error():
    messagebox.showerror("Error", "This is an error message.")

# Function to ask a question
def ask_question():
    response = messagebox.askquestion("Confirm", "Do you want to proceed?")
    if response == "yes":
        messagebox.showinfo("Response", "You selected Yes!")
    else:
        messagebox.showinfo("Response", "You selected No!")

# Add buttons to trigger message boxes
btn_info = tk.Button(window, text="Show Info", command=show_info)
btn_info.pack(pady=10)

btn_warning = tk.Button(window, text="Show Warning", command=show_warning)
btn_warning.pack(pady=10)

btn_error = tk.Button(window, text="Show Error", command=show_error)
btn_error.pack(pady=10)

btn_question = tk.Button(window, text="Ask Question", command=ask_question)
btn_question.pack(pady=10)

# Run the application
window.mainloop()
```
