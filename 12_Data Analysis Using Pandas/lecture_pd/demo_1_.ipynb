Viewing Data Frame Using loc() and iloc()

```python
import pandas as pd

df = pd.read_csv("sample.csv", encoding="latin1")
df1 = df.loc[ : ,['QTR_ID' ,	'MONTH_ID' ,	'YEAR_ID' ,'ADDRESSLINE1' , 'SALES' , 'ORDERDATE']]
print(df1)
```

```python
import pandas as pd

df = pd.read_csv("sample.csv", encoding="latin1")
df1 = df.loc[10 :16 ,['QTR_ID' ,	'MONTH_ID' ,	'YEAR_ID' ,'ADDRESSLINE1' , 'SALES']]
print(df1)
```

```python
import pandas as pd

df = pd.read_csv("sample.csv", encoding="latin1")
df1 = df.iloc[10 :16 , 1:4]
print(df1)
```
