The following are the methods you can use to create different plots with Matplotlib sub module PyPlot:


bar(): For a bar plot

hist(): Plot a Histogram

pie(): Plot a pie chart

scatter(): Form a scatter plot

stem(): Form a stem plot

step(): Form a step plot

Let us now see an example wherein we will use the Matplotlib module PyPlot and draw:

```python
import matplotlib.pyplot as plt
import numpy as np
 
# x for x-coodinate
# y for y-coordinate
x = np.array([10, 87])
y = np.array([20, 78])
 

# Plot using the pyplot.plot() method
plt.plot(x, y)
 
# Display the figure
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt   

# Read the JSON file
dd = pd.read_json('sample_Data.json')

# Display the DataFrame to check its structure (optional)
# print(dd.to_string())

# Plot using the correct variable name
plt.plot(dd['name'], dd['id'])
plt.xlabel('Name')
plt.ylabel('ID')
plt.title('Name vs ID')
plt.show()
```
