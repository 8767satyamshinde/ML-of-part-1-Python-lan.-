### Control Statements in Python

---

1. **Control Statements**
   - Overview of control statements and their purpose in altering the flow of a Python program.

2. **The `if` Statement**
   - Example and explanation:
   ```python
   age = 18
   if age >= 18:
       print("You are eligible to vote.")
   ```
   - **Problem**: Check if a number is positive or negative.
   - **Solution**:
   ```python
   number = int(input("Enter a number: "))
   if number >= 0:
       print("Positive number")
   else:
       print("Negative number")
   ```

3. **A Word on Indentation**
   - Explanation of how indentation defines code blocks in Python.
   - **Problem**: Identify which parts of the following code block belong to the `if` statement.
   ```python
   if True:
       print("Inside if block")
       print("Still inside if block")
   print("Outside if block")
   ```

4. **The `if ... else` Statement**
   - Example:
   ```python
   marks = 40
   if marks >= 50:
       print("Pass")
   else:
       print("Fail")
   ```
   - **Problem**: Check if a number is even or odd.
   - **Solution**:
   ```python
   number = int(input("Enter a number: "))
   if number % 2 == 0:
       print("Even")
   else:
       print("Odd")
   ```

5. **The `if ... elif ... else` Statement**
   - Example:
   ```python
   score = 85
   if score >= 90:
       print("Grade A")
   elif score >= 75:
       print("Grade B")
   else:
       print("Grade C")
   ```
   - **Problem**: Determine the type of triangle based on the lengths of its sides.
   - **Solution**:
   ```python
   a, b, c = map(int, input("Enter three sides of the triangle: ").split())
   if a == b == c:
       print("Equilateral Triangle")
   elif a == b or b == c or c == a:
       print("Isosceles Triangle")
   else:
       print("Scalene Triangle")
   ```

6. **The `while` Loop**
   - Example:
   ```python
   count = 1
   while count <= 5:
       print(count)
       count += 1
   ```
   - **Problem**: Print the sum of all numbers from 1 to a given number.
   - **Solution**:
   ```python
   n = int(input("Enter a number: "))
   sum = 0
   i = 1
   while i <= n:
       sum += i
       i += 1
   print(f"Sum of numbers from 1 to {n}: {sum}")
   ```

7. **The `for` Loop**
   - Example:
   ```python
   for i in range(5):
       print("Hello, World!")
   ```
   - **Problem**: Print the factorial of a given number.
   - **Solution**:
   ```python
   n = int(input("Enter a number: "))
   factorial = 1
   for i in range(1, n + 1):
       factorial *= i
   print(f"Factorial of {n} is {factorial}")
   ```

8. **Infinite Loops**
   - Explanation of what infinite loops are and how to avoid them.
   - **Problem**: Identify the problem in the following code that leads to an infinite loop:
   ```python
   i = 0
   while i < 5:
       print(i)
   ```

9. **Nested Loops**
   - Example:
   ```python
   for i in range(3):
       for j in range(3):
           print(i, j)
   ```
   - **Problem**: Print a multiplication table.
   - **Solution**:
   ```python
   for i in range(1, 11):
       for j in range(1, 11):
           print(i * j, end="\t")
       print()
   ```

10. **The `break` Statement**
    - Example:
    ```python
    for i in range(10):
        if i == 5:
            break
        print(i)
    ```

11. **The `continue` Statement**
    - Example:
    ```python
    for i in range(5):
        if i == 2:
            continue
        print(i)
    ```

12. **The `pass` Statement**
    - Example:
    ```python
    if True:
        pass  # Placeholder statement, does nothing
    print("Code executed")
    ```

13. **The `return` Statement**
    - Example:
    ```python
    def add(a, b):
        return a + b
    print(add(3, 4))
    ```
