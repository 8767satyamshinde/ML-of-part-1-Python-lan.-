```python
import  matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
print(sns.get_dataset_names())
```

```python
sns.displot(data=iris, x='sepal_length', kde=True)
```

```python
sns.displot(data=iris, x='sepal_length', kde=False)
```

```python
sns.displot(data=iris, x='sepal_length', bins=4)
```

```python
sns.displot(data=iris, x='sepal_length', kind='kde')
```

```python
sns.displot(data=iris, x='sepal_length', kind='ecdf')
```
