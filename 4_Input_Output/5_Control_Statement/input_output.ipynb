### IPython Notebook Outline for Python Syntax Guide

---

1. **Output Statements**
   - Overview of output statements in Python and their importance in displaying information.

2. **The `print()` Statement**
   - Basic usage of the `print()` statement without any arguments.
   ```python
   print()  # Prints a blank line
   ```

3. **The `print("string")` Statement**
   - Printing strings directly.
   ```python
   print("Hello, World!")  # Output: Hello, World!
   ```

4. **The `print(variables list)` Statement**
   - Printing multiple variables in a single `print` statement.
   ```python
   a, b, c = 1, 2, 3
   print(a, b, c)  # Output: 1 2 3
   ```

5. **The `print(object)` Statement**
   - Demonstrating printing Python objects like lists, dictionaries, or class instances.
   ```python
   my_list = [1, 2, 3]
   print(my_list)  # Output: [1, 2, 3]
   ```

6. **The `print("string", variables list)` Statement**
   - Printing a combination of strings and variables.
   ```python
   name, age = "Alice", 25
   print("Name:", name, "Age:", age)  # Output: Name: Alice Age: 25
   ```

7. **The `print(formatted string)` Statement**
   - Using formatted strings (`f-strings`) to format output.
   ```python
   name, age = "Alice", 25
   print(f"Name: {name}, Age: {age}")  # Output: Name: Alice, Age: 25
   ```

8. **Input Statements**
   - Using `input()` to accept user input.
   ```python
   user_name = input("Enter your name: ")
   print(f"Hello, {user_name}!")
   ```

9. **Accepting Multiple Inputs in the Same Line**
   - Accepting multiple inputs using `input().split()`.
   ```python
   x, y = input("Enter two numbers separated by space: ").split()
   print(f"First number: {x}, Second number: {y}")
   ```

10. **Command Line Arguments**
    - Explanation of how to use command line arguments using the `sys` module.
    ```python
    # This script needs to be run from the command line.
    import sys
    print("Script name:", sys.argv[0])
    if len(sys.argv) > 1:
        print("Arguments passed:", sys.argv[1:])
    ```
